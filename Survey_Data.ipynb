# Which Station to Go Next?

```python
print(survey.columns)
survey.columns = ['Grab Table?', 'Which stations did you get food from today?',
       'Family Meal', 'Global Market', 'Grill', 'Pizza', 'Special',
       'Gluten Free', 'Vegan', 'Kosher', 'Soups', 'Salad Bar', 'Water/Juices',
       'Drinks', 'Deli', 'Fruit', 'Bread', 'Desserts', 'Leave line?',
       'If Yes, which station(s)?', 'More food?', 'If Yes, which stations?']

print(survey.shape)

survey_seq = survey[['Family Meal', 'Global Market', 'Grill', 'Pizza', 'Special',
       'Gluten Free', 'Vegan', 'Kosher', 'Soups', 'Salad Bar', 'Water/Juices',
       'Drinks', 'Deli', 'Fruit', 'Bread', 'Desserts']]

print(survey_seq.shape)

survey_seq = survey_seq.reset_index()

survey_seq.columns = ['Index', 'Family Meal', 'Global Market', 'Grill', 'Pizza', 'Special',
       'Gluten Free', 'Vegan', 'Kosher', 'Soups', 'Salad Bar', 'Water/Juices',
       'Drinks', 'Deli', 'Fruit', 'Bread', 'Desserts']

# melt
survey_seq_melt = pd.melt(frame=survey_seq, id_vars = 'Index', var_name = 'Station', value_name = 'Order')
```

```python
survey_seq_melt
```

```python
survey_seq_melt = survey_seq_melt.dropna()
survey_seq_melt
```

```python
arrays = [survey_seq_melt['Index'].values, survey_seq_melt.Order.values]
tuples = list(zip(*arrays))
tuples

index = pd.MultiIndex.from_tuples(tuples, names=['Index', 'Order'])

survey_seq_melt.index = index
survey_seq_melt.index
survey_seq_melt.drop(['Index', 'Order'], axis=1, inplace=True)
survey_seq_melt

survey_seq_transform = survey_seq_melt.unstack(level = -1)
```

## Move from One Station to Next

```python
prob['Second'] = prob['Second'].fillna('Table')

second = prob.groupby(['First', 'Second']).size().to_frame('freq').reset_index()

denominator = prob.groupby(['First']).size().to_frame('denominator').reset_index()
second = pd.merge(second, denominator, how = 'inner', on = 'First')

second['prob'] = second['freq']/second['denominator']
second
```

## Leave Line? Which Stations?

```python
survey.head()
```

```python
survey_seq_transform.head()
```

```python
leaveline = pd.concat([survey_seq_transform, survey[['Leave line?','If Yes, which station(s)?']]], axis = 1)

sum_stations = leaveline.groupby(leaveline[1]).size().to_frame('freq_'+str(1)).reset_index()
for i in range(2,14):
    tmp = leaveline.groupby(leaveline[i]).size().to_frame('freq_'+str(i)).reset_index()
    sum_stations = pd.merge(sum_stations,
                            tmp,
                            how = 'left',
                            left_on = 1.0,
                            right_on = i)
    sum_stations.drop(i, axis=1, inplace = True)

sum_stations['Total Visits'] = np.sum(sum_stations.iloc[:,2:13], axis = 1)

sum_stations.columns = [ 'Station',       'freq_1',       'freq_2',       'freq_3',
             'freq_4',       'freq_5',       'freq_6',       'freq_7',
             'freq_8',       'freq_9',      'freq_10',      'freq_11',
            'freq_12',      'freq_13', 'Total Visits']

leave = pd.DataFrame(leaveline['If Yes, which station(s)?'].str.split(','))
leave['If Yes, which station(s)?'] = leave['If Yes, which station(s)?'].astype(str)

leave_stations = pd.DataFrame(columns = ['Station', 'Leave Counts'])
leave_stations['Station'] = ['Family Meal', 'Global Market', 'Grill', 'Pizza', 'Special',
       'Gluten Free', 'Vegan', 'Kosher', 'Soups', 'Salad Bar', 'Water/Juices',
       'Drinks', 'Deli', 'Fruit', 'Bread', 'Desserts']

for i in ['Family Meal', 'Global Market', 'Grill', 'Pizza', 'Special',
       'Gluten Free', 'Vegan', 'Kosher', 'Soups', 'Salad Bar', 'Water/Juices',
       'Drinks', 'Deli', 'Fruit', 'Bread', 'Desserts']:
    leave_stations.loc[leave_stations['Station']==i, 'Leave Counts'] = leave['If Yes, which station(s)?'].str.contains(i).sum()

leave_stations

leave_prob = pd.merge(sum_stations, leave_stations, how = 'inner', on = 'Station')

leave_prob['Leave prob'] = leave_prob['Leave Counts']/(leave_prob['Total Visits'] + leave_prob['Leave Counts'])
```

```python
refill = pd.DataFrame(survey['If Yes, which stations?'].str.split(','))

refill['If Yes, which stations?'] = refill['If Yes, which stations?'].astype(str)

refill_stations = pd.DataFrame(columns = ['Station', 'Refill Counts'])
refill_stations['Station'] = ['Family Meal', 'Global Market', 'Grill', 'Pizza', 'Special',
       'Gluten Free', 'Vegan', 'Kosher', 'Soups', 'Salad Bar', 'Water/Juices',
       'Drinks', 'Deli', 'Fruit', 'Bread', 'Desserts']

for i in ['Family Meal', 'Global Market', 'Grill', 'Pizza', 'Special',
       'Gluten Free', 'Vegan', 'Kosher', 'Soups', 'Salad Bar', 'Water/Juices',
       'Drinks', 'Deli', 'Fruit', 'Bread', 'Desserts']:
    refill_stations.loc[refill_stations['Station']==i, 'Refill Counts'] = refill['If Yes, which stations?'].str.contains(i).sum()
```

```python
refill_stations['Refill Sum'] = refill_stations['Refill Counts'].sum()
```

```python
refill_stations['Refill Prob'] = refill_stations['Refill Counts']/refill_stations['Refill Sum']
```
