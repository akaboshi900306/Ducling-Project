```python
dic = {
    'Water/Juices': 1,
    'Family Meal': 2,
    'Global Market': 3,
    'Grill': 4,
    'Vegan': 5,
    'Kosher': 6,
    'Gluten Free': 7,
    'Soups': 8,
    'Drinks': 9,
    'Deli': 10,
    'Salad Bar': 11,
    'Special': 12,
    'Pizza': 13,
    'Fruit': 14,
    'Desserts': 15,
    'Bread': 16
}
```

```python
import numpy as np
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import os
os.chdir(
    r'C:\Users\HSANG3\Desktop\MSBA 2019\Fall\ISOM 677 Decision Analytic Optimization\Project Ducling'
)

data = pd.read_csv(r'data\Survey Data\Cleaned Data\Ducling_Survey_Cleaned.csv')
data = data.replace('No', 0)
data = data.replace('Yes', 1)
data = data.replace({"1st Destination": dic})
data = data.replace({"2nd Destination": dic})
data = data.replace({"3rd Destination": dic})
for i in range(4, 14):
    data = data.replace({(str(i) + "th Destination"): dic})
data = data.fillna(0)
data
```

```python
order = data.iloc[:, 2:-4]
index = ('Water/Juices', 'Family Meal', 'Global Market', 'Grill', 'Vegan',
         'Kosher', 'Gluten Free', 'Soups', 'Drinks', 'Deli', 'Salad Bar',
         'Special', 'Pizza', 'Fruit', 'Desserts', 'Bread')
columns = ('Water/Juices', 'Family Meal', 'Global Market', 'Grill', 'Vegan',
           'Kosher', 'Gluten Free', 'Soups', 'Drinks', 'Deli', 'Salad Bar',
           'Special', 'Pizza', 'Fruit', 'Desserts', 'Bread', 'Exit')

count = pd.DataFrame(0, index=index, columns=columns)
ratio = pd.DataFrame(0, index=index, columns=columns)

for row in range(0, 256):
    for col in range(0, 12):
        i = int(order.iloc[row, col])
        j = int(order.iloc[row, col + 1])
        if (i != 0 and (j == 0 or col == 12)):
            count.iloc[i - 1, 16] += 1
        elif j != 0:
            count.iloc[i - 1, j - 1] += 1
count

for i in range(0, 16):
    for j in range(0, 17):
        ratio.iloc[i, j] = round(count.iloc[i, j] / count.iloc[i].sum() * 100,
                                 2)
ratio

sum(count['Exit'])
ratio.sum(axis=1)
```

```python
import csv

ratio.to_csv('ratio.csv', sep=',', na_rep="NA", header=True, index=True)
```
